# **For Loops with Lists**



```
for item in items:
    # indented block – executed once per item
```



## **How It Works**

1. First item of the list is **assigned** to the variable item.

2. The indented block runs, using that variable.

3. Python moves to the next item and repeats.

4. Loop ends after the last item.

**Example:**

````python
fruits = ["Apple", "Peach", "Pear"]
for fruit in fruits:
    print(fruit)
    print(fruit + " pie")
````

## **Indentation – The “Inside / Outside” Rule**

1. **Colon (:)** – marks the beginning of a loop (or conditional) block.
2. **Indented lines** → **inside** the loop – executed every **iteration.**
3. **Dedented lines** → outside the loop – executed once after the loop finishes.

**Note:** Indentation is Python’s way of grouping statements.
Misplacing indentation changes the program’s logic silently.



## **The Accumulator Pattern**

**Definition:** A variable that is updated (accumulated) inside a loop to build a final result.





### **Summation – Manual vs Built‑in**

````python
student_scores = [150, 142, 185, 120, 171, 184, 149, 199]

# Manual accumulator
total = 0
for score in student_scores:
    total += score          # add current score to total
print(total)

# Built-in
print(sum(student_scores))
````

### **Maximum Value – Manual vs Built‑in**

````python
# Manual approach
max_score = 0
for score in student_scores:
    if score > max_score:
        max_score = score
print(max_score)

# Built-in
print(max(student_scores))
````

**Note:** The manual maximum works only if **all scores are positive.**

# **For Loops with range()**
Used when you know **how many times to repeat**, not necessarily with an existing list.



Screenshot_2026-02-11_21-57-48.png

**Key rule: The stop value is never included.**

**Example: Print 1 to 10**

````python
for number in range(1, 11):
  print(number)
````

**Gauss Challenge – Sum 1 to 100**

````python
total_sum = 0

for number in range(1, 101):
  total_sum += number
print(total_sum)
````

**This works:** The loop runs 100 times; each iteration adds the current number to total_sum

## **FizzBuzz – Problem Statement**

Rules:

Print numbers from 1 to 100 (inclusive).

Divisible by 3 → print "Fizz" instead of the number.

Divisible by 5 → print "Buzz" instead.

Divisible by both 3 and 5 → print "FizzBuzz".

````python
for number in range(1, 101):
  if number % 3 and number % 5 == 0:
    print("FizzBuzz")
  elif number % 3 == 0:
    print("Fizz")
  elif number % 5 == 0:
    print("Buzz")
  else:
    print(number)
````
