```python
words_list = [
    "action", "aspect", "backup", "basket", "bottle",
    "bridge", "bright", "broken", "budget", "butter",
    "camera", "castle", "change", "choice", "coffee",
    "column", "damage", "danger", "design", "dinner",
    "doctor", "energy", "escape", "family", "figure",
    "flower", "forest", "friend", "garden", "ground",
    "guitar", "handle", "health", "island", "jacket",
    "league", "market", "memory", "museum", "nature",
    "office", "orange", "parent", "player", "pocket",
    "police", "public", "school", "search", "silver"
]
```

```python
guess = input("Guess a letter: ").lower()
print(guess)
```

```python
import random

chosen_word = random.choice(words_list)

print(chosen_word)
```

```python
for i in chosen_word:
  if i == guess:
    print("Right")
  elif i != guess:
    print("Wrong")
  return
```
