```python
import random
```

```python
user_choice = int(input("Type 0 for Rock, 1 for Paper, 2 for Scissors: "))
```

```python
computer_choice = random.randint(0, 2)
print(computer_choice)
```

```python
if user_choice == 0 and computer_choice == 0:
    print("It's a Tie!")
elif user_choice == 0 and computer_choice == 1:
    print("Computer Wins!")  # Paper covers Rock
elif user_choice == 0 and computer_choice == 2:
    print("User Wins!")  # Rock smashes Scissors
elif user_choice == 1 and computer_choice == 1:
    print("It's a Tie!")
elif user_choice == 1 and computer_choice == 0:
    print("User Wins!")  # Paper covers Rock
elif user_choice == 1 and computer_choice == 2:
    print("Computer Wins!")  # Scissors cut Paper
elif user_choice == 2 and computer_choice == 2:
    print("It's a Tie!")
elif user_choice == 2 and computer_choice == 1:
    print("User Wins!")  # Scissors cut Paper
elif user_choice == 2 and computer_choice == 0:
    print("Computer Wins!")  # Rock smashes Scissors
```
