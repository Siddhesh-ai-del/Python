# **1. Randomization in Python**


## **1.1 Core Concepts**
- Randomization: Introducing unpredictability in programs (e.g., games).

- Deterministic Computers: Execute repeatable, predictable operations.

- Pseudo-Random Number Generators (PRNGs): Algorithms that generate seemingly random numbers using mathematical formulas.

## **1.5 Common Application: Coin Toss Program**

````python
coin = random.randint(0, 1)

if coin == 0:
  print("Head!")
elif coin == 1:
  print("Tails!")
else:
  pass
````

# **2. Python Modules**


## **2.1 What is a Module?**
- A Python file containing related code (functions, variables) that can be imported and reused.
- Purpose: Organize code, enable collaboration, separate functionality.

## **2.2 Creating & Using Custom Modules**

1. Create my_module.py:


```
# my_module.py
my_favorite_number = 3.1415
```



2. Import in main file:


```
import my_module
print(my_module.my_favorite_number)
```
