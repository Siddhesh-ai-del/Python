```python
print("""
Welcome to Treasure Island.
Your mission is to find the treasure.
You're at a cross road. Where do you want to go?
""")

cross_road = input('Type "left" or "right" ')
if cross_road == "right":
  print("You fell into a hole. Game Over")
elif cross_road == "left":
  print("You have come to a lake. There is an island in the middle of the lake")

  lake_choice = input('Type "wait" to wait for a boat. Type "swim" to swim across. ')
  if lake_choice == "swim":
    print("You get attacked by a sea monster. Game Over.")
  elif lake_choice == "wait":
    print("""
    You arrive at the island unharmed. There is a house with 3 doors.
    One red, one yellow and one blue.
    """)
    doors_choice = input("Which colour do you choose? ")

    if doors_choice == "red":
      print("It's a room full of fire. Game Over.")
    elif doors_choice == "yellow":
      print("You found the treasure! You Win!")
    elif doors_choice == "blue":
      print("You enter a room of beasts. Game Over.")
    else:
      pass
else:
  pass
```
