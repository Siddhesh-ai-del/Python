# **Basic if / else Statement**

```
if condition:
    # Code block if condition is True
else:
    # Code block if condition is False
```
**Remember!**
  1. Colon (:) after if and else.

  2. Indentation defines code blocks (typically 4 spaces). Incorrect indentation causes IndentationError.

  3. else is optional. It catches all cases where the if condition was False.

````python
height = int(input("What is your height in cm? "))
if height >= 120:
    print("You can ride the roller coaster.")
else:
    print("Sorry, you have to grow taller before you can ride.")
````

# **The Modulo Operator (%)**

**Definition & Purpose**

1. Modulo (%) is a mathematical operator that returns the remainder after division.

2. Purpose: Crucial for checking divisibility (e.g., odd/even numbers).

````python
number = int(input("Which number do you want to check? "))
if number % 2 == 0:
    print("This is an even number.")
else:
    print("This is an odd number.")
````

# **elif for Multiple Exclusive Conditions**

elif = "else if". Checks another condition only if the previous if/elif was False.


```
if condition1:
    # Block A
elif condition2:
    # Block B
else:
    # Block C (if all above are False)
```



````python
age = int(input("What is your age? "))
if age < 12:
    bill = 5
elif age <= 18:  # Caught if age is 12 to 18 inclusive
    bill = 7
else:
    bill = 12
print(f"Your ticket costs ${bill}.")
````

# **Nested if Statements**
1. An if/elif/else statement placed inside another if block.
2. Logic: The inner condition is only checked if the outer condition passed.

````python
if height >= 120:
    print("You can ride.")
    age = int(input("What is your age? "))
    if age <= 18:
        print("Please pay $7.")
    else:
        print("Please pay $12.")
else:
    print("You cannot ride.")
````

````python
# First, determine base ticket price (only one runs)
if age < 12:
    bill = 5
elif age <= 18:
    bill = 7
else:
    bill = 12

# Then, an INDEPENDENT check for an optional add-on
wants_photo = input("Do you want a photo? (Y/N) ")
if wants_photo == "Y":
    bill += 3  # Add $3 to the existing bill

print(f"Your final bill is ${bill}.")
````
