# **Tip Calculator Project**

1. Gather the Inputs

```python
bill = input("What was the total bill? $")
tip_percent = input("What percentage tip would you like to give? ")
people = input("How many people to split the bill? ")
```

2. Make The Logic fo Tip calculator

```python
bill_float = float(bill)        # Bill can have decimals
tip_int = int(tip_percent)      # Tip is a whole number
people_int = int(people)        # Can't have fractional people

tip_multiplier = 1 + (tip_int / 100)
total_with_tip = bill_float * tip_multiplier

per_person = total_with_tip / people_int
final_amount = round(per_person, 2)  # Round to 2 decimal places
```

3. Display the result

```python
print(f"Each person should pay: ${final_amount}")
```
