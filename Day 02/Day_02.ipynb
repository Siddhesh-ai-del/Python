# **Data Type:**
A classification of data which tells the compiler/interpreter how the programmer intends to use the data.


 Python functions and operations behave differently depending on the data type provided. Using the wrong type causes **TypeError.**


1. **String (str)**	Sequence of characters (text).
2. **Integer (int**)	Whole numbers (positive, negative, or zero).
3. **Float (float)**	Numbers with decimal points ("floating-point").
4. **Boolean (bool)**	Logical values representing **True or False.**

**Note!** "123" is a string, 123 is an integer. Their behavior with operators differs fundamentally.

# **String Indexing & Subscripting**

**Subscripting:** Accessing individual characters using square brackets [].

**Indexing starts at 0**. The first character is at index 0.

**Positive Indices:** Count from the left (start).

**Negative Indices:** Count from the right (end). **-1 is the last character.**

```python
name = "Siddhesh"

print(name[0])   # Output: S
print(name[4])   # Output: h
print(name[-1])  # Output: h (Last character)
```

**Common Misunderstanding:** Forgetting indexing starts at 0, leading to "off-by-one" errors.

# **String Concatenation**

The + operator joins (concatenates) strings.

**It does NOT perform mathematical addition on numeric strings.**

```python
print("123" + "456") # Output: "123456" (Concatenation)
print(123 + 456) # Output: 579 (Mathematical addition)
```

# **Checking Type: type() function**

  Use type(data) to identify the data type of any variable.

  

```python
print(type("Hello")) # <class 'str'>
print(type(123))     # <class 'int'>
print(type(3.14))    # <class 'float'>
print(type(True))    # <class 'bool'>
```

# **Type Conversion (Casting)**

1. str() function converts data to **string**


*   Example: str(123) → "123"
*   Caution: it Always works.

2. int() converts data to **Integer**

*   Example: int("123") → 123 & int(3.99) → 3 (truncates)
*   Caution: Fails for non-numeric strings: int("ABC") causes ValueError.

3. float() converts data to **Float**

*   Example: float("3.14") → 3.14 & float(3) → 3.0

4. bool() converts data to Boolean

*   Example: bool(1) → True & bool(0) → False
*   Caution: 0, "" (empty string), None are **False.** Most other values are **True.**









# **Common Type Error & Fix**

1. Error: TypeError: can only concatenate str (not "int") to str
2. Cause: Trying to use + between a string and a non-string (e.g., int).
3. Fix: Convert the non-string to a string first.


```python
# ERROR: print("Your score is " + 100)
# FIX:
print("Your score is " + str(100))
```

# **Order of Operations (PEMDAS/LR)**
**Rule:** Parentheses → Exponents → Multiplication & Division → Addition & Subtraction.

**Left-to-Right (LR):** For operators with equal precedence (like * and /), calculate from left to right.

```python
print(3 * 3 + 3 / 3 - 3)  # Calculation order: (3*3)=9, (3/3)=1, 9+1=10, 10-3=7.0
```

```python
print(3 * (3 + 3) / 3 - 3) # Order: (3+3)=6, 3*6=18, 18/3=6, 6-3=3.0
```

# **Rounding Numbers: round()**

1. **round(number, digits)** rounds a number to a specified precision.
2. **number:** The value to round.
3. **digits (optional):** Number of decimal places (default is 0).

```python
bmi = 30.85125
print(round(bmi))      # 31 (to nearest whole number)
print(round(bmi, 2))  # 30.85 (to 2 decimal places)
```

**⚠️ Distinction:**

  **int(30.85) → 30** (flooring, chops off decimal).

  **round(30.85) → 31** (mathematical rounding).

# **F-Strings for Clean Formatting**

1. Problem: Manually converting and concatenating multiple data types to print is cumbersome.
2. Solution: f-Strings (formatted string literals).


*   **Syntax: f"Your text {variable}"**
*   Place variables or expressions inside curly braces {}. Automatic type conversion occurs.





```python
score = 0
height = 1.8
is_winning = True
print(f"Your score is {score}, your height is {height}, you are winning is {is_winning}")
```
