## **Maze Project – Follow the Right Edge**

**Goal:** Navigate any maze, robot starts at random position & direction.

**Algorithm (from World Info):**
Follow the right‑hand wall – i.e., always try to turn right; if can’t, go straight; if can’t go straight, turn left.


```
def turn_right():
    turn_left()
    turn_left()
    turn_left()

while not at_goal():
    if right_is_clear():
        turn_right()
        move()
    elif front_is_clear():
        move()
    else:
        turn_left()
```
